```python
def readfiles(path1, path2):
    list1 = []
    list2 = []
    with open(path1, "r") as file: 
        ln = file.readline() 
        ln = file.readline()
        while ln: 
            delim = ln.split('\t')
            a = delim[1][9:]
            list1.append(a)
            ln = file.readline()
            
    with open(path2, "r") as file2: 
        ln = file2.readline() 
        ln = file2.readline()
        while ln: 
            delim = ln.split('\t')
            a = delim[1][9:]
            list2.append(a)
            ln = file2.readline()  
    
    return list1,list2
            
def compare(path1, path2): 
    list1 = readfiles(path1, path2)[0]
    list2 = readfiles(path1, path2)[1]
    a = (set(list1) & set(list2))
    for i in a: 
        print(i)
    return a
    
def papercomp(a,b):
    b = b.split('\n')
    return set(a).intersection(b)
```
